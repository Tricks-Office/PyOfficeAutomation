```python
from docx import Document
import pandas as pd
import os
import re

# 읽어올 파일 경로 및 폴더 경로 지정 TPath : 타겟 파일 경로, RPath : 결과 파일 경로, 
#                                 DPath : 사전 파일 경로
TPath = r'C:\Data\Sample\4_4_Change_Words_Word\Data'
RPath = r'C:\Data\Sample\4_4_Change_Words_Word\Result'
DPath = r'C:\Data\Sample\4_4_Change_Words_Word\Dictionary.xlsx'

# 읽어올 폴더에서 파일 이름 리스트 가져오기
docfiles = [f for f in os.listdir(TPath) if re.match(r'.*[.]doc', f)]

# 사전 정보 읽어오기
df = pd.read_excel(DPath)

for docfile in docfiles:
    # 워드 파일 읽어오기
    doc = Document(os.path.join(TPath,docfile))

    # 찾을 단어에 대해서 반복
    for index, dfrow in df.iterrows():
        # 각 문단에서 반복문 실행
        for paragraph in doc.paragraphs:
            # 문단의 각 run(띄어쓰기) 단위로 반복문 실행
            for run in paragraph.runs:
                # 찾을 단어가 있을 때만 실행    
                if run.text.find(dfrow['Old']) != -1 :
                    # 원하는 단어 찾아 바꾸기
                    run.text = run.text.replace(dfrow['Old'],dfrow['New'])

        # 각 표에서 반복문 실행
        for table in doc.tables:
            # 표의 각 행단위로 반복문 실행
            for row in table.rows:
                # 행 내부의 Cell 단위로 반복문 실행
                for cell in row.cells:
                    # 원하는 단어 찾아 바꾸기
                    cell.text = cell.text.replace(dfrow['Old'],dfrow['New'])

    doc.save(os.path.join(RPath,docfile))
```
