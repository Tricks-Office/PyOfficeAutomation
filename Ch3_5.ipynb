### SUMIF 함수 구현하기

```python
import pandas as pd

df1 = pd.read_excel('C:/Data/Sample/3_5_Python_Excel/매출관리Sample_Mod.xlsx')
df2 = pd.read_excel('C:/Data/Sample/3_5_Python_Excel/매출관리Sample_Mod.xlsx', sheet_name = 1)
df3 = pd.read_excel('C:/Data/Sample/3_5_Python_Excel/매출관리Sample_Mod.xlsx', sheet_name = 2)

dfr = df1.merge(df2, how='left', left_on = '담당자 코드', right_on='사번')
dfr = dfr.merge(df3, how='left', on = '강의코드')

dfr['매출액'] = dfr['단가']*dfr['수강인원']
```

```python
dfg = dfr[['이름','수강인원','매출액']].groupby('이름').sum()
```

```python
dfg=dfr[['이름','수강인원','매출액']].groupby('이름').count()
```
